```python
##iterating over dict
##you can use loops to iterate over dictionaries, keys, values or items

##iterating over keys
for key in student.keys():
    print(key)
```

```python
for value in student.values():
    print(value)
```

```python
#iterate over key value pairs
for key, value in student.items():
    print(f"{key}:{value}")
```

```python
##nested dictionaries
students={
    "student1":{'name':'suresh','age':23},
    "student2":{'name':'peter','age':21}
}
print(students)
```

```python
##acess nested dict
print(students['student2']['name'])
print(students['student2']['age'])
```

```python
students.items()
```

```python
#iterating over nested dict
for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
#dict comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
##conditional dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
##practical examples

##use a dictionary to count the freq of elements in the list
numbers=[1,2,2,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
#merge dict

dict1={'a':1,'b':2}
dict2={'b':3,'c':4}
merge_dict={**dict1,**dict2}
print(merge_dict)
```
