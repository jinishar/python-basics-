```python
#slicing list
numbers=[1,2,3,4,5,6,7,8]
print(numbers)
print(numbers[:1])
print(numbers[:5])
print(numbers[-1])
print(numbers[2:5])
print(numbers[::2])
print(numbers[::-1])
print(numbers[:])

```

```python
#iterate over list

for number in numbers:
    print(number)
```

```python
##iterate with index
for index,number in enumerate (numbers): #enumrate gives you the access to the index of each values
    print(index,number)
```

```python
[x**2 for x in range(10)]
```

#### List comprehension
basic syntax       [expression for item iteratble]
with conditional logic  [expression for item in iteratble if condition]
nested list comprehension [expression for item1 in iterable1 for item2 in iterable2]

```python
#basic list comprhension
square = [num**2 for num in range(10)]
print(square)
```

```python
#basic list comprhension with condition
lst=[]
for i in range(10):
    if i%2==0:
        lst.append(i)

print(lst)
```

```python
even_numbers=[num for num in range(10) if num%2==0]
print(even_numbers)
```
