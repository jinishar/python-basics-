```python
def sq(x):
    return x*x

sq(4)
```

```python
numbers=[1,2,3,4,5,6,7,8]
list(map(sq,numbers))
```

```python
##lambda func w map
numbers=[1,2,3,4,5,6,7,8]

list(map(lambda x:x*x,numbers))
```

```python
### map multiple iterables
numbers1=[1,2,3]
numbers2=[4,5,6]

added_num=list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_num)
```

```python
#map() convert a list of stringa to integers
#use map to convert strings to integers
str1=['1','2','3','4','5']
int1=list(map(int, str1))   #inbuilt int 
print(int1)
```

```python
words=['apple','banana','cherry']
upper_word=list(map(str.upper,words))
print(upper_word)
```

```python
def get_name(person):
    return person['name']

people=[
    {'name':'john','age':32},
    {'name':'doe','age':44}
]
list(map(get_name,people))
```
