```python
#simple function
def addition(a,b):
    return a+b

print(addition(6,7))
```

```python
#using lambda function
add=lambda a,b:a+b
type(add)
print(add(4,5))
```

```python
def even(num):
    if num%2==0:
        return True
even(24)
```

```python
value= lambda num:num%2==0
value(40)
```

```python
def add(x,y,z):
    return x+y+z
add(2,3,5)
```

```python
addition=lambda x,y,z:x+y+z
print(addition(12,14,16))
```

```python
##map() -  it applies a func to all items in the list
numbers=[1,2,3,4,5,6]
def sq(number):
    return number**2

sq(3)
```

```python
list(map(lambda x:x**2,numbers))
```
